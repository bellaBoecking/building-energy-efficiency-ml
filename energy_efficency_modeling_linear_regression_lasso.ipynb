# Energy Efficency Modeling Using Linear Regression and Lasso

##### Project Overview
The purpose of this project is to model building energy efficiency based on building parameters, using statistical learning techniques to model heating and cooling loads. We aim to construct interpretable, computationally efficient models while maintaining strong predictive performance through linear regression and Lasso regression.

##### Motivation
Effective building energy efficiency modeling allows for better-informed decisions in early design phases, helping to reduce energy consumption and overhead costs.

##### Dataset Information
The dataset, sourced from the UCI Machine Learning Repository, provides a total of 768 instances with eight predictors. It contains two response variables, the heating load and the cooling load, both of which describe building energy efficiency.

##### Process Overview
After preprocessing, including scaling and encoding categorical variables, we begin by establishing a linear regression baseline, evaluating residuals and mean squared errors (MSE) to observe and address issues of non-linearity and heteroscedasticity. We first conduct response variable transformations before changing the model to better accommodate non-linearity. We then introduce interaction terms, using Lasso with cross-validation to improve model performance and maintain model simplicity. Final model evaluations are performed with R^2 values, cross-validated MSE, and residual plots from a validation set.

##### Project Outcome
The project demonstrates a complete machine learning pipeline, including preprocessing, model selection, feature engineering, hyperparameter tuning, and model evaluation. This project yielded a computationally efficient, relatively simple model that effectively predicts heating and cooling loads based on building parameters. We achieved strong predictive performances, with R^2 values of ~.96 and ~.93 for the heating load and cooling load, respectively.

### Library Imports

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import ElasticNetCV
import sklearn.model_selection as skm
from sklearn.model_selection import KFold, ShuffleSplit
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model as skl
from sklearn.metrics import r2_score
```

### Loading Data

```python
df = pd.read_excel("ENB2012_data.xlsx")
```

### Proprocessing

```python
# Changing column names for easy interpretability
df.columns = ['Relative Compactness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Orientation', 'Glazing Area', 'Glazing Area Distribution', 'Heating Y', 'Cooling Y']
df
```

```python
# We'll one-hot encode the categorical variables and drop the first dummy in each category
df_encoded = pd.get_dummies(df, columns = ['Orientation', 'Glazing Area Distribution'], drop_first = True)
df_encoded
```

The output label Y1 is the heating load while Y2 is the cooling load, each target variable is regressed seperately.

```python
Y1 = df['Heating Y']
Y2 = df['Cooling Y']
X = df_encoded.drop(columns = ['Heating Y', 'Cooling Y'])
```

### Linear Regression

We begin with linear regression to establish a baseline before trying more complex methods if necessary.

```python
K = 10
kfold = skm.KFold(K, 
                  random_state = 0,
                  shuffle = True)
```

```python
# We want to scale only the numerical variables, not the categorical ones, so we use the sklearn.compose ColumnTransformer
numerical_cols = ['Relative Compactness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Glazing Area']
# 'Standardize' is the name of the transformer we apply
# 'passthrough' tells ColumnTransformer to let untouched columns stay
preprocessor = ColumnTransformer([('Standardize', StandardScaler(), numerical_cols)], remainder = 'passthrough')                 
```

```python
# The pipeline does fit_transform for us
pipeline = Pipeline(steps = [('preprocessing', preprocessor), ('linreg', LinearRegression())]) 
```

Here we'll create a function that performs linear regression with K-fold cross-validation and returns both the predictions and the MSE scores. We create our own function because cross_val returns either the predictions or the MSE scores for each fold, not both.

```python
def lin_reg_cross_val(kfold, X, Y, pipeline, mse_type):
    mse_scores = []
    predictions = np.zeros(len(Y)) # np.zeros(length)
    val_idx_lst = []

    for train_idx, val_idx in kfold.split(X):
        X_train, Y_train = X.iloc[train_idx], Y.iloc[train_idx] # Here we use iloc to index bc X, Y are df, and the [] notation
        # tries to access a column, whereas for numpy it accesses a row
        X_test, Y_test = X.iloc[val_idx], Y.iloc[val_idx]
        
        pipeline.fit(X_train, Y_train)
        Y_pred = pipeline.predict(X_test) # Returns a numpy array

        predictions[val_idx] = Y_pred

        if mse_type == 'sqrt':
            Y_pred_original = Y_pred ** 2
            Y_test_original = Y_test.to_numpy() ** 2 # convert to numpy to avoid index errors
            mse = ((Y_pred_original - Y_test_original) ** 2).mean()
        elif mse_type == 'log':
            mse = ((np.exp(Y_pred) - np.exp(Y_test.to_numpy()))**2).mean()
        else:
            mse = ((Y_pred - Y_test.to_numpy())**2).mean()
            
        mse_scores.append(mse)
        val_idx_lst.extend(val_idx)

    val_idx_lst = np.array(val_idx_lst)
    sort_order = np.argsort(val_idx_lst)
    val_idx_lst = val_idx_lst[sort_order]
    predictions = predictions[sort_order] # Sorting the predictions in the original order of Y
    
    return mse_scores, predictions
```

Now we'll apply the function to yeild the results for both the models.

```python
heat_Y_mse, heat_Y_pred = lin_reg_cross_val(kfold, X, Y1, pipeline, 'reg')
cool_Y_mse, cool_Y_pred = lin_reg_cross_val(kfold, X, Y2, pipeline, 'reg')

heat_Y_mse_mean = np.mean(heat_Y_mse) # .mean() only works for numpy arrays, so we convert them here to get the mean easily
cool_Y_mse_mean = np.mean(cool_Y_mse)
print("MSE mean of heating model:", heat_Y_mse_mean)
print("MSE mean of cooling model:", cool_Y_mse_mean)
```

Let's compare the linear regression MSE to that of a naive model which simply takes the mean of the responce.

```python
def naive_mean(kfold, X, Y):
    mse_scores = []
    predictions = np.zeros(len(Y))

    for train_idx, val_idx in kfold.split(X):
        X_train, Y_train = X.iloc[train_idx], Y.iloc[train_idx]
        X_test, Y_test = X.iloc[val_idx], Y.iloc[val_idx]

        Y_pred = np.ones(len(Y_test)) * np.mean(Y_train) # Multiplies each element

        predictions[val_idx] = Y_pred
        mse = ((Y_pred - Y_test.to_numpy())**2).mean()
        mse_scores.append(mse)

    return mse_scores, predictions
```

```python
naive_heat_mse_mean = np.mean(naive_mean(kfold, X, Y1)[0])
naive_cool_mse_mean = np.mean(naive_mean(kfold, X, Y2)[0])

print("Naive heat MSE mean:", naive_heat_mse_mean)
print("Naive cool MSE mean:", naive_cool_mse_mean)
```

Already, the linear regression model is performing much better than taking the mean of the response variables. We'll plot the residuals of the linear regression model to look for patterns that might indicate we need a more flexible model or if we should try other, less flexible linear models. 

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 8))
ax1.scatter(heat_Y_pred, Y1 - heat_Y_pred)
ax2.scatter(cool_Y_pred, Y2 - cool_Y_pred)

ax1.set_ylim(-40, 40)
ax2.set_ylim(-40, 40)
ax1.set_xlim(0, 45)
ax2.set_xlim(0, 45)

ax1.set_xlabel("Fitted Value")
ax1.set_ylabel("Residual")
ax2.set_xlabel("Fitted Value")
ax2.set_ylabel("Residual")
ax1.axhline(0, c = 'k', ls = '--')
ax2.axhline(0, c = 'k', ls = '--')
plt.suptitle("Linear Regression Residual Plot");
ax1.set_title("Heating Load")
ax2.set_title("Cooling Load");
```

These residuals are interesting because they exhibit a non-uniform spread around y = 0. The pattern appears to be linear, and the model is systematically overestimating small values of Y and underestimating large values of Y. Thus, we'll try some adjustments to the target variable to see if they "flatten" out our tilted cloud.

#### Target Transformation

We'll start with adjustments to the target variable to stabalize the variance of the residuals. We went back to add another argument to lin_reg_cross_val, mse_type, to scale the MSE correctly so we can fairly compare it between models.

```python
Y1_log = np.log(Y1)
Y2_log = np.log(Y2)

heat_Y_mse, heat_Y_log_pred = lin_reg_cross_val(kfold, X, Y1_log, pipeline, 'log')
cool_Y_mse, cool_Y_log_pred = lin_reg_cross_val(kfold, X, Y2_log, pipeline, 'log')

heat_Y_mse_mean = np.mean(heat_Y_mse)
cool_Y_mse_mean = np.mean(cool_Y_mse)

print("Heat Y MSE: ", heat_Y_mse_mean)
print("Cooling Y MSE: ", cool_Y_mse_mean)
```

```python
Y1_log = np.sqrt(Y1)
Y2_log = np.sqrt(Y2)

heat_Y_mse, heat_Y_sqrt_pred = lin_reg_cross_val(kfold, X, Y1_log, pipeline, 'sqrt')
cool_Y_mse, cool_Y_sqrt_pred = lin_reg_cross_val(kfold, X, Y2_log, pipeline, 'sqrt')

heat_Y_mse_mean = np.mean(heat_Y_mse)
cool_Y_mse_mean = np.mean(cool_Y_mse)

print("Heat Y MSE: ", heat_Y_mse_mean)
print("Cooling Y MSE: ", cool_Y_mse_mean)
```

Changes to the target variable improved the MSE of both models for both types of changes. However, the improvements are modest. Let's take a look at the residuals to see if the new models better accommodate the heteroscedasticity of the data.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 8))

# adjusting the predictions to original scale
adj_heat_Y_log_pred = np.exp(heat_Y_log_pred)
adj_cool_Y_log_pred = np.exp(cool_Y_log_pred)

ax1.scatter(adj_heat_Y_log_pred, Y1 - adj_heat_Y_log_pred)
ax2.scatter(adj_cool_Y_log_pred, Y2 - adj_cool_Y_log_pred)

ax1.set_ylim(-40, 40)
ax2.set_ylim(-40, 40)
ax1.set_xlim(0, 50)
ax2.set_xlim(0, 50)

ax1.set_xlabel("Fitted Value")
ax1.set_ylabel("Residual")
ax2.set_xlabel("Fitted Value")
ax2.set_ylabel("Residual")
ax1.axhline(0, c = 'k', ls = '--')
ax2.axhline(0, c = 'k', ls = '--')
fig.suptitle("Log(Y) Linear Regression Residual Plot");
ax1.set_title("Heating Load")
ax2.set_title("Cooling Load");
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 8))

# adjusting the predictions to original scale
adj_heat_Y_sqrt_pred = heat_Y_sqrt_pred ** 2
adj_cool_Y_sqrt_pred = cool_Y_sqrt_pred ** 2

ax1.scatter(adj_heat_Y_sqrt_pred, Y1 - adj_heat_Y_sqrt_pred)
ax2.scatter(adj_cool_Y_sqrt_pred, Y2 - adj_cool_Y_sqrt_pred)

ax1.set_ylim(-40, 40)
ax2.set_ylim(-40, 40)
ax1.set_xlim(0, 50)
ax2.set_xlim(0, 50)

ax1.set_xlabel("Fitted Value")
ax1.set_ylabel("Residual")
ax2.set_xlabel("Fitted Value")
ax2.set_ylabel("Residual")
ax1.axhline(0, c = 'k', ls = '--')
ax2.axhline(0, c = 'k', ls = '--')
fig.suptitle("Sqrt(Y) Linear Regression Residual Plot");
ax1.set_title("Heating Load")
ax2.set_title("Cooling Load");
```

The model still doesn't capture the data's heteroscedasticity as effectively as we'd like it to. The shape of the residual graph points to a mispecified conditional mean--the relationship between x and y isn't strictly linear. If the issue were solely heteroscedasticity, one might expect to see more improvement with the adjustments to the target variable performed earlier. The model's modest improvement indicates that adjustments to the model's underlying structure are needed. We'll take a look at the data to better observe variable relationships before changing the model.

### Plotting Variables

```python
sns.pairplot(df)
plt.show()
```

Here we see some highly correlated variable relationships, such as that of surface area and relative compactness, while others show little to no relationship at all, such as that of orientation plotted against all other variables. Performing Lasso with interaction terms is the next step in our process because it allows the machine to remove unnecessary terms (such as orientation) while accommodating for potential nonlinearity as indicated by the residual behavior. 

### Interaction Term Lasso

Here we want a model that performs lasso with variable "options" that include the original features as well as squared interaction terms. We construct the polynomial features and pipeline accordingly. We'll leave out a section of the dataset to validate with later to observe residuals.

```python
col_indicies = list(range(6))
# We have to use column indices here because what PolynomialFeatures returns is a NumPy array, which CT will try to access as a df
post_scale = ColumnTransformer([('scale', StandardScaler(), col_indicies)], remainder = 'passthrough')
```

```python
def validation_set_split(X, Y):
    
    validation = ShuffleSplit(n_splits = 1,
                              test_size = .2,
                              random_state = 0)
    for train_idx, val_idx in validation.split(X): # We use iloc because X, Y are df and [] attempts to access columns
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        Y_train, Y_val = Y.iloc[train_idx], Y.iloc[val_idx]

    return X_train, X_val, Y_train, Y_val
```

#### Heating Load Lasso

```python
X_train, X_val, Y_train, Y_val = validation_set_split(X, Y1)

lassoCV = skl.ElasticNetCV(n_alphas = 100, # ElasticNetCV already gives us an intercept
                         l1_ratio = 1,
                         cv = kfold,
                         max_iter = 10000)

# Note that we already have a bias term
pipe = Pipeline([('preprocessor', preprocessor),
                 ('poly', PolynomialFeatures(degree = 2, interaction_only = True, include_bias = False)),
                 ('post_scale', post_scale),
                ('lasso', lassoCV)])

heat_fit = pipe.fit(X_train, Y_train)
```

```python
tuned_lasso = heat_fit.named_steps['lasso']
tuned_lasso.alpha_
```

This represents the alpha corresponding to the lowest MSE.

```python
lassoCV_fig, ax = plt.subplots(figsize = (8, 8))
ax.errorbar(-np.log(tuned_lasso.alphas_), 
            tuned_lasso.mse_path_.mean(1),
            yerr = tuned_lasso.mse_path_.std(1) / np.sqrt(K))
ax.axvline(-np.log(tuned_lasso.alpha_), c = 'k', ls = '--')
ax.set_xlabel('$-\log(\lambda)$')
ax.set_ylabel('Cross-validated MSE')
ax.set_title("Heat Load Lasso Cross-Validated MSE vs. Log(Î»)");
```

```python
count_nonzero = sum(1 for i in tuned_lasso.coef_ if i != 0)
print(tuned_lasso.coef_, count_nonzero)
```

```python
tuned_lasso.mse_path_.mean(1)[-1]
```

```python
pred = pipe.predict(X_val)
fig, ax = plt.subplots(figsize = (8, 8))
ax.scatter(pred, Y_val - pred)
ax.set_xlabel("Fitted Value")
ax.set_ylabel("Residual")
ax.axhline(0, c = 'k', ls = '--')
ax.set_title("Heating Load Lasso Residuals");
```

Here, the residuals are much noisier, meaning the lasso model captures the complexities of the data much more effectively than linear regression. Furthermore, the lasso's MSE on the test set was ~4.4, much better than the corresponding score of linear regression, ~7.2 at its best, and far superior to that of the naive mean model, ~102.

```python
mse = ((Y_val - pred)**2).mean()
r2 = r2_score(Y_val, pred)
print("R^2 Score: ", r2)
print("MSE Score ", mse)
```

#### Cooling Load Lasso

Now we'll repeat the steps to verify that the same type of model also performs well on the cooling load target variable. 

```python
X_train, X_val, Y_train, Y_val = validation_set_split(X, Y2)
cool_fit = pipe.fit(X_train, Y_train)
tuned_lasso = cool_fit.named_steps['lasso']
tuned_lasso.alpha_
```

This represents the alpha corresponding to the lowest MSE.

```python
lassoCV_fig, ax = plt.subplots(figsize = (8, 8))
ax.errorbar(-np.log(tuned_lasso.alphas_), 
            tuned_lasso.mse_path_.mean(1),
            yerr = tuned_lasso.mse_path_.std(1) / np.sqrt(K))
ax.axvline(-np.log(tuned_lasso.alpha_), c = 'k', ls = '--')
ax.set_xlabel('$-\log(\lambda)$')
ax.set_ylabel('Cross-validated MSE')
ax.set_title("Cool Load Lasso Cross-Validated MSE vs. Log(Î»)");
```

```python
count_nonzero = sum(1 for i in tuned_lasso.coef_ if i != 0)
print(tuned_lasso.coef_, count_nonzero)
```

```python
tuned_lasso.mse_path_.mean(1)[-1]
```

```python
pred = pipe.predict(X_val)
fig, ax = plt.subplots(figsize = (8, 8))
ax.scatter(pred, Y_val - pred)
ax.set_xlabel("Fitted Value")
ax.set_ylabel("Residual")
ax.axhline(0, c = 'k', ls = '--');
ax.set_title("Cooling Load Lasso Residuals");
```

```python
mse = ((Y_val - pred)**2).mean()
r2 = r2_score(Y_val, pred)
print("R^2 Score: ", r2)
print("MSE Score ", mse)
```

Lasso with interaction terms appears to work very will with the cooling load target variable as well, yeilding a ~6.1 MSE at its best and an R2 score of ~.93. This is slightly lower than that of the heating load model but very successful nonetheless.

### Evaluation and Conclusions

This project utilized linear regression, target transformation, and lasso with interaction terms to model building energy efficiency based on building parameters such as glazing area, relative compactness, roof area, and more, with a total of eight features. Linear regression was chosen as a baseline due to its interpretability, simplicity, and clear residual diagnostics. Ultimately, it was unable to capture nonlinear relationships in the data and thus exhibited a non-uniform distribution of residuals. The MSEs of linear regression for the heating load and cooling load were ~8.3 and ~10.3, respectively; a significant improvement compared to the naive model, which simply predicted the mean of the training targets and yielded MSEs of ~102.1 and ~90.7.

The distribution of the linear regression's residuals took the shape of a tilted parallelogram, indicating a systematic overestimation of low response variable values and underestimation of high response variable values. Addressing heteroscedasticity first, we decided to transform the target variable with log(Y) and sqrt(Y) to stabilize variance. This strategy yielded slight improvements, which made it clear that variance wasn't the model's core issue. 

Given the shape of the residuals and the modest improvements yielded through target transformation, we introduced nonlinearity through Lasso with interaction terms. Lasso was an attractive option for this dataset because of the patterns observed in the plotting variables section. While some of the variables were highly correlated, others, such as orientation, showed little to no correlation with other variables. Furthermore, based on subject knowledge, potential interactions between variables such as relative compactness and glazing area were observed. Lasso accommodated these properties of the data by setting unnecessary predictors to zero, allowing the model to drop some of the regular predictors as well as unhelpful interaction terms to yield a simpler model. 

Lasso with interaction terms performed very well on the data, resulting in MSEs of ~4.4 and ~6.1 and R1 scores of ~.96 and ~.93 for the respective heating and cooling load models. Furthermore, it effectively captured the relationships and variance in the data as the corresponding residual graphs were noisy and showed no clear pattern. Overall, this project yielded a computationally efficient, relatively simple model that effectively predicts heating and cooling loads based on building parameters.
